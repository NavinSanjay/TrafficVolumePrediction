## Automate going through all the pages

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
df = pd.DataFrame()
weather_data = []
```

```python
browser = webdriver.Chrome(executable_path=r'C:\chromedriver\chromedriver.exe')
```

```python
month = 6
year = 2017
website = f"https://www.timeanddate.com/weather/new-zealand/auckland/historic?month={month}&year={year}"
```

```python
browser.get(website)
```

```python
select_element = Select(browser.find_element(By.ID, "wt-his-select"))

# Get all the options from the <select> element
options = select_element.options
print(options)
# Iterate over each option and scroll to it, then click
for option in options:
    print("option start: ", option)
    # Scroll to the option
    browser.execute_script("arguments[0].scrollIntoView(true);", option)

    # Wait for the option to be clickable
    wait = WebDriverWait(browser, 10)
    option = wait.until(EC.element_to_be_clickable((By.XPATH, "//option[text()='{}']".format(option.text))))

    # Click the option
    option.click()

    # Perform scraping or data extraction here
    
    #Table Data:
    page_soup = soup(browser.page_source,"html.parser")
    table = page_soup.find_all(("table",{"id":"wt-his"}))
    print(table)
    table_rows = table[1].find_all('tr')
    table_rows = table_rows[2:]
    print("rows:",table_rows)
#---------------------------------------
    #Table Scrape
    # Create a list to store the weather data
    i = 0
    print(len(table_rows))
    # Iterate over the table rows and extract the data
    for row in table_rows:
        if i == ((len(table_rows)) - 1): #Stopping Condition. When at the last row of the Table
            break
        else:
            print(row)        
            try:

                time_date = row.find_all("th")[0]
                time_date = str(time_date)
                print(time_date)
                # Extract the time and Date
                time_regex = re.compile(r'>(.*?)<br/>') #Tries for the normal rows first
                time_matches = re.findall(time_regex, time_date)
                time_= time_matches[0] #if time_matches else ""
                print(time_)  
            except:
                time_date = row.find_all("th")[0]
                time_date = str(time_date)
                print(time_date)
                # Extract the time and Date
                time_regex = re.compile(r'>(.*?)</th>') #Tries for the normal rows first
                time_matches = re.findall(time_regex, time_date)
                time_ = time_matches[0] #if time_matches else ""
                print(time_)


            #Date
            try:
                date_regex = re.compile(r'<span class="smaller soft">(.*?)</span>')
                date_matches = re.findall(date_regex, time_date)
                date = date_matches[0]
                print(date)
            except: #Except when passed the first row. Then dont need the date anymore
                date = date
                print(date)

            # Extract the table cells from the row
            cells = row.find_all('td')

            # Extract the data from the cells
            # time = cells[0].text.strip()
            # print(time) 
            temperature = cells[1].text.strip()
            print(temperature)
            weather = cells[2].text.strip()
            print(weather)
            wind = cells[3].text.strip()
            print(wind)
            humidity = cells[5].text.strip()
            print(humidity)
            barometer = cells[6].text.strip()
            print(barometer)

            # Append the data to the weather_data list
            weather_data.append([time_, date, temperature, weather, wind, humidity, barometer])
            print(i)
            i += 1

    # Create a DataFrame from the weather_data list
    df = pd.DataFrame(weather_data, columns=['Time','Date', 'Temperature', 'Weather', 'Wind', 'Humidity', 'Barometer'])
    print("option end: ", option)
    # print(df)
    
# Close the WebDriver
browser.quit()
```
