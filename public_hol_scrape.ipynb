```python
# List to store the extracted data
data = []
```

```python

# Iterate through each table
i = 0 
for table in tables:
    year = h2_tags[i].text
    print("Table Header:", year)
    
    # Iterate through rows and columns to extract the data
    for row in table.find_all("tr"):
        cells = row.find_all("td")
        row_data = [cell.text.strip() for cell in cells]
        row_data.insert(0, year)  # Insert the year at the beginning of the row data
        data.append(row_data)
    i += 1
```

```python
df = pd.DataFrame(data, columns=["Year", "Public Holiday", "Actual Date", "Observed Date"])
```
